### Loads up the data

```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv("../diamonds.csv")
df_test  = pd.read_csv("../diamonds_new.csv")
```

```python
display(df_train.head())
display(df_test.head())
```

```python
### Drops a few attributes
df_train.drop('Unnamed: 0',axis=1, inplace=True)
df_test.drop('Unnamed: 0',axis=1, inplace=True)

df_train.drop('cut',axis=1, inplace=True)
df_test.drop('cut',axis=1, inplace=True)

df_train.drop('color',axis=1, inplace=True)
df_test.drop('color',axis=1, inplace=True)

df_train.drop('clarity',axis=1, inplace=True)
df_test.drop('clarity',axis=1, inplace=True)
```

```python
display(df_train.head())
display(df_test.head())
```

### Creating the linear model

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
X_train  = df_train.drop('price', axis = 1)
Y_train  = df_train['price']
X_test   = df_test
### Y_predict = ?

### fits model
linreg = SGDRegressor(fit_intercept=True, random_state=15)

linreg.fit(X_train, Y_train)
coefs = linreg.coef_
intercept = linreg.intercept_[0]
print intercept

predictions_test = linreg.predict(X_test)

# predictions_train = linreg.predict(X_train)
# score = r2_score(predictions_train, Y_train)
# print score
```

### Prints the equation

```python
index = 0
equation = "Price = "
for att in df_test.columns.values:
    coef = coefs[index]
    equation = equation + "{:.2f} * {} + ".format(coef, att)
    index = index + 1
equation = equation + "{:.2f}".format(intercept)
print equation
```

### Step 1

According to the model, if a diamond is 1 carat heavier than another with the same cut, how much more should I expect to pay? Why?


```python
carat1 = [1, 0, 0]
carat2 = [2, 0, 0]
v1 = linreg.predict(carat1)
v2 = linreg.predict(carat2)
print "{} to {}".format(v1, v2)
```

If you were interested in a 1.5 carat diamond with a Very Good cut (represented by a 3 in the model) and a VS2 clarity rating (represented by a 5 in the model), how much would the model predict you should pay for it?


```python
t1 = [1.5, 3, 5]
vt1 = linreg.predict(t1)
print "{:.2f}".format(vt1[0])
```

```python
df_test_labeled = df_test.copy()
df_test_labeled['price'] = predictions_test
df_test_labeled.plot(kind='scatter', x='carat', y='price', color='red')

negatives = df_test_labeled.loc(df_test_labeled['price'] < 0)
display(negatives)
```

### Computing the recommendation

```python
margin = 0.7 * predictions_test
bidding = margin.sum()
print "Bidding: {:.2f}".format(bidding)
```
